```python
import sys
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from graphviz import Source
from sklearn.model_selection import train_test_split, cross_val_score, cross_validate
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, BaggingClassifier, RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import os
from sklearn.model_selection import GridSearchCV, RepeatedStratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import ComplementNB
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import pickle

enable_plot_tree = False
enable_plot_confusion_matrix = True

# Load dataset
dataset = pd.read_csv('ai4i2020.csv')

# DATA ANALYSIS:
## dimensionality reduction
columns_to_drop = ["Product ID", "UDI", "TWF", "HDF", "PWF", "OSF", "RNF"]
dataset.drop(columns_to_drop, axis=1, inplace=True)

# Remove non-informative duplicates
pd.DataFrame.drop_duplicates(dataset, subset=dataset.columns.difference(['Machine failure']), inplace=True)

# One hot encoding of categorical feature
one_hot = pd.get_dummies(dataset['Type'])
dataset = dataset.drop('Type', axis=1)
last = dataset.iloc[:, -1]
dataset = dataset.drop('Machine failure', axis=1)
dataset = dataset.join(one_hot)
dataset = dataset.join(last)

# Create images directory
PROJECT_ROOT_DIR = "."
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images")
os.makedirs(IMAGES_PATH, exist_ok=True)

def plot_confusion_matrix_wrapper(model, X_test, y_test, title):
    """Wrapper function to plot confusion matrix"""
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.title(title)
    plt.show()

    # Print classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

def print_feature_importance(model, feature_names):
    """Print feature importance if the model has it"""
    if hasattr(model, 'feature_importances_'):
        print("\nFeature Importance:")
        importance = model.feature_importances_
        for i, (name, score) in enumerate(zip(feature_names, importance)):
            print(f"{i+1}. {name}: {score:.4f}")
    elif hasattr(model, 'coef_'):
        print("\nFeature Coefficients:")
        coef = model.coef_[0]
        for i, (name, score) in enumerate(zip(feature_names, coef)):
            print(f"{i+1}. {name}: {score:.4f}")

def train_test_split_standard_scaler(X, Y, train_size=0.65, random_state=42):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=train_size, random_state=random_state)
    scaler = StandardScaler()
    scaler.fit(X_train)
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)
    return (X_train, X_test, Y_train, Y_test)

def evaluate_model(model, X_train, X_test, Y_train, Y_test, feature_names, model_name):
    """Generic function to evaluate any model"""
    print(f"\n=== {model_name.upper()} ===")

    # Cross-validation
    cv_results = cross_validate(model, X_train, Y_train, cv=5, n_jobs=-1,
                               scoring=['recall', 'accuracy'], return_estimator=True)

    # Get the best estimator (you might want to change this logic)
    best_estimator = cv_results['estimator'][0]

    # Test set evaluation
    Y_pred = best_estimator.predict(X_test)

    # Confusion matrix
    cm = confusion_matrix(Y_test, Y_pred)
    tn, fp, fn, tp = cm.ravel()

    print("\nTest Set Results:")
    print(f"True Negative: {tn}")
    print(f"False Positive: {fp}")
    print(f"False Negative: {fn}")
    print(f"True Positive: {tp}")

    # Plot confusion matrix
    if enable_plot_confusion_matrix:
        plot_confusion_matrix_wrapper(best_estimator, X_test, Y_test, f'Confusion Matrix - {model_name}')

    # Feature importance
    print_feature_importance(best_estimator, feature_names)

    return best_estimator

def decision_treeWeighted(dataset):
    dataset = dataset.drop(['L', 'M', 'H'], axis=1)
    X = dataset.iloc[:, :-1]
    Y = dataset.iloc[:, -1]
    feature_names = X.columns.tolist()

    X_train, X_test, Y_train, Y_test = train_test_split_standard_scaler(X, Y)

    model = DecisionTreeClassifier(max_depth=4, class_weight='balanced',
                                 criterion='entropy', random_state=42,
                                 min_samples_leaf=30)

    return evaluate_model(model, X_train, X_test, Y_train, Y_test, feature_names, "Decision Tree (Weighted)")

def svm_weighted(dataset):
    dataset = dataset.drop(['L', 'M', 'H'], axis=1)
    X = dataset.iloc[:, :-1]
    Y = dataset.iloc[:, -1]
    feature_names = X.columns.tolist()

    X_train, X_test, Y_train, Y_test = train_test_split_standard_scaler(X, Y)

    model = SVC(kernel='rbf', C=float('inf'), gamma=1,
               class_weight='balanced', random_state=42)

    return evaluate_model(model, X_train, X_test, Y_train, Y_test, feature_names, "SVM (Weighted)")

def svm_sampling(dataset):
    dataset = dataset.drop(['L', 'M', 'H'], axis=1)
    X = dataset.iloc[:, :-1]
    Y = dataset.iloc[:, -1]
    feature_names = X.columns.tolist()

    X_train, X_test, Y_train, Y_test = train_test_split_standard_scaler(X, Y)

    # Apply sampling
    over = SMOTE(sampling_strategy=0.45, random_state=42)
    under = RandomUnderSampler(sampling_strategy=0.5, random_state=42)
    X_train, Y_train = over.fit_resample(X_train, Y_train)
    X_train, Y_train = under.fit_resample(X_train, Y_train)

    model = SVC(kernel='rbf', C=float('inf'), gamma=1,
               class_weight='balanced', random_state=42)

    return evaluate_model(model, X_train, X_test, Y_train, Y_test, feature_names, "SVM (Sampling)")

def decisionTree_sampling(dataset):
    dataset = dataset.drop(['L', 'M', 'H'], axis=1)
    X = dataset.iloc[:, :-1]
    Y = dataset.iloc[:, -1]
    feature_names = X.columns.tolist()

    X_train, X_test, Y_train, Y_test = train_test_split_standard_scaler(X, Y)

    # Apply sampling
    over = SMOTE(sampling_strategy=0.45, random_state=42)
    under = RandomUnderSampler(sampling_strategy=0.5, random_state=42)
    X_train, Y_train = over.fit_resample(X_train, Y_train)
    X_train, Y_train = under.fit_resample(X_train, Y_train)

    model = DecisionTreeClassifier(criterion='entropy', max_depth=4,
                                 min_samples_leaf=30, class_weight='balanced')

    best_model = evaluate_model(model, X_train, X_test, Y_train, Y_test, feature_names, "Decision Tree (Sampling)")

    if enable_plot_tree:
        draw_tree("tree-sampling", best_model, dataset)

    return best_model

def randomForest_weighted(dataset):
    dataset = dataset.drop(['L', 'M', 'H'], axis=1)
    X = dataset.iloc[:, :-1]
    Y = dataset.iloc[:, -1]
    feature_names = X.columns.tolist()

    X_train, X_test, Y_train, Y_test = train_test_split_standard_scaler(X, Y)

    model = RandomForestClassifier(n_jobs=-1, criterion='entropy', n_estimators=1000,
                                 class_weight='balanced_subsample', random_state=42,
                                 max_samples=0.5, max_leaf_nodes=16)

    return evaluate_model(model, X_train, X_test, Y_train, Y_test, feature_names, "Random Forest (Weighted)")

def samples_randomForest(dataset):
    dataset = dataset.drop(['L', 'M', 'H'], axis=1)
    X = dataset.iloc[:, :-1]
    Y = dataset.iloc[:, -1]
    feature_names = X.columns.tolist()

    X_train, X_test, Y_train, Y_test = train_test_split_standard_scaler(X, Y)

    # Apply sampling
    over = SMOTE(sampling_strategy=0.45, random_state=42)
    under = RandomUnderSampler(sampling_strategy=0.5, random_state=42)
    X_train, Y_train = over.fit_resample(X_train, Y_train)
    X_train, Y_train = under.fit_resample(X_train, Y_train)

    model = RandomForestClassifier(n_jobs=-1, criterion='entropy', n_estimators=1000,
                                 random_state=42, max_samples=0.5, max_leaf_nodes=16,
                                 class_weight='balanced_subsample')

    return evaluate_model(model, X_train, X_test, Y_train, Y_test, feature_names, "Random Forest (Sampling)")

def decisionTreeWithMostImportantFeature(dataset):
    dataset = dataset.drop(['L', 'M', 'H', 'Air temperature [K]', 'Process temperature [K]'], axis=1)
    X = dataset.iloc[:, :-1]
    Y = dataset.iloc[:, -1]
    feature_names = X.columns.tolist()

    X_train, X_test, Y_train, Y_test = train_test_split_standard_scaler(X, Y)

    model = DecisionTreeClassifier(max_depth=4, class_weight='balanced',
                                 criterion='entropy', random_state=42,
                                 min_samples_leaf=30)

    best_model = evaluate_model(model, X_train, X_test, Y_train, Y_test, feature_names,
                              "Decision Tree (Most Important Features)")

    if enable_plot_tree:
        draw_tree("most-important", best_model, dataset)

    return best_model

def gaussian_naive(dataset):
    dataset = dataset.drop(['L', 'M', 'H'], axis=1)
    X = dataset.iloc[:, :-1]
    Y = dataset.iloc[:, -1]
    feature_names = X.columns.tolist()

    X_train, X_test, Y_train, Y_test = train_test_split_standard_scaler(X, Y)

    model = GaussianNB()

    return evaluate_model(model, X_train, X_test, Y_train, Y_test, feature_names, "Gaussian Naive Bayes")

def gaussian_naive_sampling(dataset):
    dataset = dataset.drop(['L', 'M', 'H'], axis=1)
    X = dataset.iloc[:, :-1]
    Y = dataset.iloc[:, -1]
    feature_names = X.columns.tolist()

    X_train, X_test, Y_train, Y_test = train_test_split_standard_scaler(X, Y)

    # Apply sampling
    over = SMOTE(sampling_strategy=0.45, random_state=42)
    under = RandomUnderSampler(sampling_strategy=0.5, random_state=42)
    X_train, Y_train = over.fit_resample(X_train, Y_train)
    X_train, Y_train = under.fit_resample(X_train, Y_train)

    model = GaussianNB()

    return evaluate_model(model, X_train, X_test, Y_train, Y_test, feature_names, "Gaussian Naive Bayes (Sampling)")

def reteNeurale(dataset):
    dataset = dataset.drop(['L', 'M', 'H'], axis=1)
    X = dataset.iloc[:, :-1]
    Y = dataset.iloc[:, -1]
    feature_names = X.columns.tolist()

    X_train, X_test, Y_train, Y_test = train_test_split_standard_scaler(X, Y, train_size=0.75)

    # Apply sampling
    over = SMOTE(sampling_strategy=0.45, random_state=42)
    under = RandomUnderSampler(sampling_strategy=0.5, random_state=42)
    X_train, Y_train = over.fit_resample(X_train, Y_train)
    X_train, Y_train = under.fit_resample(X_train, Y_train)

    model = MLPClassifier(solver='adam', validation_fraction=0.25, early_stopping=True,
                         activation='relu', learning_rate='adaptive', n_iter_no_change=10,
                         random_state=42, alpha=1e-6, max_iter=500)

    return evaluate_model(model, X_train, X_test, Y_train, Y_test, feature_names, "Neural Network")

# Main execution
if __name__ == "__main__":
    # List of all model functions to evaluate
    model_functions = [
        decision_treeWeighted,
        svm_weighted,
        svm_sampling,
        decisionTree_sampling,
        randomForest_weighted,
        samples_randomForest,
        decisionTreeWithMostImportantFeature,
        gaussian_naive,
        gaussian_naive_sampling,
        reteNeurale
    ]

    # Evaluate all models
    trained_models = {}
    for model_func in model_functions:
        model_name = model_func.__name__
        print(f"\n{'='*50}")
        print(f"EVALUATING {model_name.upper()}")
        print(f"{'='*50}")

        trained_model = model_func(dataset)
        trained_models[model_name] = trained_model
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, cross_validate
from sklearn.ensemble import RandomForestClassifier, IsolationForest
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (confusion_matrix, ConfusionMatrixDisplay,
                           classification_report, precision_recall_curve,
                           fbeta_score, make_scorer)
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler, TomekLinks
from imblearn.pipeline import Pipeline
from xgboost import XGBClassifier
import matplotlib.pyplot as plt

# Load and prepare data
def load_and_preprocess_data():
    dataset = pd.read_csv('ai4i2020.csv')

    # Drop unnecessary columns
    cols_to_drop = ["Product ID", "UDI", "TWF", "HDF", "PWF", "OSF", "RNF"]
    dataset.drop(cols_to_drop, axis=1, inplace=True)

    # Remove duplicates
    dataset.drop_duplicates(subset=dataset.columns.difference(['Machine failure']), inplace=True)

    # One-hot encoding
    one_hot = pd.get_dummies(dataset['Type'])
    dataset = dataset.drop('Type', axis=1)
    last = dataset.iloc[:, -1]
    dataset = dataset.drop('Machine failure', axis=1)
    dataset = dataset.join(one_hot)
    dataset = dataset.join(last)

    # Feature engineering
    dataset['Torque_x_RPM'] = dataset['Torque [Nm]'] * dataset['Rotational speed [rpm]']
    dataset['ToolWear_x_Torque'] = dataset['Tool wear [min]'] * dataset['Torque [Nm]']

    return dataset

# Advanced sampling techniques
def get_sampling_pipeline(strategy='smote_tomek'):
    if strategy == 'smote_tomek':
        over = SMOTE(sampling_strategy=0.5, random_state=42)
        under = TomekLinks()
    elif strategy == 'adasyn':
        over = ADASYN(sampling_strategy=0.5, random_state=42)
        under = RandomUnderSampler(sampling_strategy=0.8, random_state=42)
    else:  # default
        over = SMOTE(sampling_strategy=0.5, random_state=42)
        under = RandomUnderSampler(sampling_strategy=0.8, random_state=42)

    return Pipeline([('over', over), ('under', under)])

# Model training with threshold optimization
def train_model_with_threshold(model, X_train, y_train, X_test, y_test, threshold=0.5):
    model.fit(X_train, y_train)
    y_proba = model.predict_proba(X_test)[:, 1]
    y_pred = (y_proba >= threshold).astype(int)

    # Metrics
    cm = confusion_matrix(y_test, y_pred)
    print(classification_report(y_test, y_pred))
    print(f"F2-Score: {fbeta_score(y_test, y_pred, beta=2):.4f}")

    # Plot confusion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.title(f'Confusion Matrix (Threshold={threshold})')
    plt.show()

    # Precision-Recall curve
    precision, recall, thresholds = precision_recall_curve(y_test, y_proba)
    plt.plot(recall, precision)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.show()

    return model, y_proba

# Main function
def main():
    # Load and preprocess data
    dataset = load_and_preprocess_data()
    X = dataset.drop(['Machine failure', 'L', 'M', 'H'], axis=1)
    y = dataset['Machine failure']

    # Train-test split with standardization
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Sampling
    sampler = get_sampling_pipeline(strategy='smote_tomek')
    X_train_res, y_train_res = sampler.fit_resample(X_train, y_train)

    # Models to evaluate
    models = {
        'XGBoost': XGBClassifier(
            scale_pos_weight=10,
            n_estimators=500,
            max_depth=6,
            learning_rate=0.1,
            random_state=42
        ),
        'Random Forest': RandomForestClassifier(
            n_estimators=500,
            max_depth=6,
            class_weight='balanced_subsample',
            random_state=42
        ),
        'Weighted SVM': SVC(
            kernel='rbf',
            C=1.0,
            class_weight={0:1, 1:10},
            probability=True,
            random_state=42
        ),
        'Isolation Forest': IsolationForest(
            contamination=0.03,
            random_state=42
        )
    }

    # Train and evaluate each model
    for name, model in models.items():
        print(f"\n{'='*50}")
        print(f"Evaluating {name}")
        print(f"{'='*50}")

        if name == 'Isolation Forest':
            # Anomaly detection approach
            model.fit(X_train)
            y_pred = model.predict(X_test)
            y_pred = [1 if x == -1 else 0 for x in y_pred]  # Convert anomalies to 1
            cm = confusion_matrix(y_test, y_pred)
            print(classification_report(y_test, y_pred))
            disp = ConfusionMatrixDisplay(confusion_matrix=cm)
            disp.plot()
            plt.title(f'Confusion Matrix - {name}')
            plt.show()
        else:
            # Standard classification approach
            trained_model, y_proba = train_model_with_threshold(
                model,
                X_train_res,
                y_train_res,
                X_test,
                y_test,
                threshold=0.3  # Lower threshold favors recall
            )

            # Feature importance for tree-based models
            if hasattr(trained_model, 'feature_importances_'):
                print("\nFeature Importance:")
                importance = trained_model.feature_importances_
                for i, (name, score) in enumerate(zip(X.columns, importance)):
                    print(f"{i+1}. {name}: {score:.4f}")

    # Hyperparameter tuning with F2-score
    print("\nTuning XGBoost hyperparameters...")
    f2_scorer = make_scorer(fbeta_score, beta=2)
    param_grid = {
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [4, 6, 8],
        'scale_pos_weight': [5, 10, 15]
    }
    grid = GridSearchCV(
        XGBClassifier(n_estimators=100, random_state=42),
        param_grid,
        scoring=f2_scorer,
        cv=5,
        n_jobs=-1
    )
    grid.fit(X_train_res, y_train_res)
    print(f"Best parameters: {grid.best_params_}")
    print(f"Best F2-score: {grid.best_score_:.4f}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (confusion_matrix, ConfusionMatrixDisplay,
                           classification_report, precision_recall_curve,
                           fbeta_score, make_scorer, average_precision_score)
from sklearn.preprocessing import StandardScaler, LabelEncoder
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler, TomekLinks
from imblearn.pipeline import Pipeline
from xgboost import XGBClassifier
import seaborn as sns
import re

# Configuration
RANDOM_STATE = 42
TEST_SIZE = 0.3
THRESHOLD = 0.3  # Optimal threshold for recall-oriented problems

def clean_feature_names(feature_names):
    """Clean feature names by removing special characters"""
    cleaned = []
    for name in feature_names:
        # Remove brackets and content inside them
        name = re.sub(r'\[.*?\]', '', name)
        # Remove other special characters and spaces
        name = re.sub(r'[^a-zA-Z0-9_]', '_', name)
        # Remove trailing underscores
        name = name.strip('_')
        cleaned.append(name)
    return cleaned

def load_and_preprocess_data():
    """Load and preprocess data with proper type handling"""
    dataset = pd.read_csv('ai4i2020.csv')

    # First inspect the actual column names
    print("Original columns in dataset:", dataset.columns.tolist())

    # Drop unnecessary columns (using the exact column names from the file)
    cols_to_drop = ["Product ID", "UDI", "TWF", "HDF", "PWF", "OSF", "RNF"]
    dataset.drop(cols_to_drop, axis=1, inplace=True)

    # Remove duplicates while preserving the target distribution
    dataset.drop_duplicates(inplace=True)

    # Convert target to binary (handles both string and numeric representations)
    dataset['Machine failure'] = dataset['Machine failure'].astype(int)

    # Encode the 'Type' column instead of one-hot encoding
    le = LabelEncoder()
    dataset['Type'] = le.fit_transform(dataset['Type'])
    print("Type classes:", le.classes_)

    # Clean all column names
    dataset.columns = clean_feature_names(dataset.columns)
    print("Cleaned columns:", dataset.columns.tolist())

    # Feature engineering - use the cleaned column names
    # First verify the torque column exists
    torque_col = [col for col in dataset.columns if 'Torque' in col]
    rpm_col = [col for col in dataset.columns if 'Rotational' in col or 'RPM' in col]
    wear_col = [col for col in dataset.columns if 'Tool' in col and 'wear' in col]

    if torque_col and rpm_col:
        dataset['Torque_x_RPM'] = dataset[torque_col[0]] * dataset[rpm_col[0]]
    if torque_col and wear_col:
        dataset['ToolWear_x_Torque'] = dataset[wear_col[0]] * dataset[torque_col[0]]

    return dataset

def get_sampling_pipeline(strategy='smote_tomek'):
    """Create sampling pipeline with configurable strategy"""
    if strategy == 'smote_tomek':
        over = SMOTE(sampling_strategy=0.5, random_state=RANDOM_STATE, k_neighbors=5)
        under = TomekLinks()
    elif strategy == 'adasyn':
        over = ADASYN(sampling_strategy=0.5, random_state=RANDOM_STATE, n_neighbors=5)
        under = RandomUnderSampler(sampling_strategy=0.8, random_state=RANDOM_STATE)
    else:  # default
        over = SMOTE(sampling_strategy=0.5, random_state=RANDOM_STATE)
        under = RandomUnderSampler(sampling_strategy=0.8, random_state=RANDOM_STATE)

    return Pipeline([('over', over), ('under', under)])

def evaluate_model(model, X_train, X_test, y_train, y_test, feature_names, model_name):
    """Enhanced model evaluation with comprehensive metrics"""
    # Train model
    model.fit(X_train, y_train)

    # Get probabilities and predictions
    y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else [0]*len(X_test)
    y_pred = (y_proba >= THRESHOLD).astype(int)

    # Calculate metrics
    cm = confusion_matrix(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)
    f2 = fbeta_score(y_test, y_pred, beta=2)
    avg_precision = average_precision_score(y_test, y_proba) if hasattr(model, "predict_proba") else 0

    # Print results
    print(f"\n{'='*50}")
    print(f"MODEL: {model_name.upper()}")
    print(f"{'='*50}")
    print(f"F2-Score: {f2:.4f}")
    print(f"Average Precision: {avg_precision:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name}', pad=20)
    plt.show()

    # Plot enhanced precision-recall curve if model supports probabilities
    if hasattr(model, "predict_proba"):
        precision, recall, _ = precision_recall_curve(y_test, y_proba)
        baseline = len(y_test[y_test==1]) / len(y_test)  # Random guess

        plt.figure(figsize=(8, 6))
        plt.plot(recall, precision, marker='.', label=model_name)
        plt.axhline(y=baseline, color='r', linestyle='--', label='Random Guess')
        plt.xlabel('Recall (Sensitivity)', fontsize=12)
        plt.ylabel('Precision (PPV)', fontsize=12)
        plt.title(f'Precision-Recall Curve - {model_name}', pad=20)
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()

    # Feature importance for tree-based models
    if hasattr(model, 'feature_importances_'):
        importance = pd.DataFrame({
            'Feature': feature_names,
            'Importance': model.feature_importances_
        }).sort_values('Importance', ascending=False)

        plt.figure(figsize=(10, 6))
        sns.barplot(x='Importance', y='Feature', data=importance.head(10))
        plt.title(f'Top 10 Important Features - {model_name}', pad=20)
        plt.tight_layout()
        plt.show()

        print("\nTop 10 Features by Importance:")
        print(importance.head(10).to_string(index=False))

    return {
        'model': model,
        'f2_score': f2,
        'avg_precision': avg_precision,
        'classification_report': report
    }

def main():
    # Set style for plots
    sns.set_style("whitegrid")
    plt.rcParams['figure.facecolor'] = 'white'

    # Load and preprocess data
    dataset = load_and_preprocess_data()

    # Show target distribution
    target_col = [col for col in dataset.columns if 'Machine' in col and 'failure' in col][0]
    print("Target variable distribution:")
    print(dataset[target_col].value_counts())
    print(f"\nImbalance ratio: {len(dataset[dataset[target_col]==0])/len(dataset[dataset[target_col]==1]):.1f}:1")

    # Prepare features and target
    X = dataset.drop(target_col, axis=1)
    y = dataset[target_col]
    feature_names = X.columns.tolist()

    # Split data with stratification
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=TEST_SIZE,
        random_state=RANDOM_STATE,
        stratify=y
    )

    # Standardize only numerical features
    numerical_cols = X.select_dtypes(include=['float64', 'int64']).columns
    scaler = StandardScaler()
    X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
    X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])

    # Handle class imbalance
    sampler = get_sampling_pipeline(strategy='smote_tomek')
    X_train_res, y_train_res = sampler.fit_resample(X_train, y_train)

    # Define models
    models = {
        'XGBoost': XGBClassifier(
            scale_pos_weight=10,
            n_estimators=500,
            max_depth=6,
            learning_rate=0.1,
            random_state=RANDOM_STATE,
            eval_metric='logloss'
        ),
        'Random Forest': RandomForestClassifier(
            n_estimators=500,
            max_depth=6,
            class_weight='balanced_subsample',
            random_state=RANDOM_STATE,
            n_jobs=-1
        ),
        'SVM': SVC(
            kernel='rbf',
            C=1.0,
            class_weight='balanced',
            probability=True,
            random_state=RANDOM_STATE
        ),
        'Neural Network': MLPClassifier(
            hidden_layer_sizes=(100, 50),
            early_stopping=True,
            random_state=RANDOM_STATE,
            max_iter=500
        )
    }

    # Evaluate models
    results = {}
    for name, model in models.items():
        results[name] = evaluate_model(
            model, X_train_res, X_test, y_train_res, y_test,
            feature_names, name
        )

    # Hyperparameter tuning for best model
    print("\n" + "="*60)
    print("HYPERPARAMETER TUNING FOR XGBOOST".center(60))
    print("="*60)

    param_grid = {
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 6],
        'subsample': [0.8, 1.0],
        'colsample_bytree': [0.8, 1.0],
        'scale_pos_weight': [5, 10, 15]
    }

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    grid = GridSearchCV(
        XGBClassifier(n_estimators=100, random_state=RANDOM_STATE),
        param_grid,
        scoring=make_scorer(fbeta_score, beta=2),
        cv=cv,
        n_jobs=-1,
        verbose=1
    )

    grid.fit(X_train_res, y_train_res)

    print("\nBest parameters:", grid.best_params_)
    print("Best F2-score:", grid.best_score_)

    # Evaluate best model
    best_model = grid.best_estimator_
    best_results = evaluate_model(
        best_model, X_train_res, X_test, y_train_res, y_test,
        feature_names, "Tuned XGBoost"
    )

    # Compare all models
    print("\n" + "="*60)
    print("MODEL COMPARISON".center(60))
    print("="*60)
    comparison = pd.DataFrame({
        'Model': list(results.keys()) + ['Tuned XGBoost'],
        'F2-Score': [res['f2_score'] for res in results.values()] + [best_results['f2_score']],
        'Avg Precision': [res['avg_precision'] for res in results.values()] + [best_results['avg_precision']]
    }).sort_values('F2-Score', ascending=False)

    print(comparison.to_string(index=False))

if __name__ == "__main__":
    main()
```
